```python
import yfinance as yf
import pandas as pd

ticker = ["BTC-USD", "CL=F"]
start_date="2024-01-01"
end_date="2024-12-31"

data = yf.download(ticker, start=start_date, end=end_date)
display(data.head())
```

```python
#select the 'Close' price and calculate daily returns
close_prices = data['Close']
daily_returns = close_prices.pct_change()*100

#Drop rows with NaN values
daily_returns = daily_returns.dropna()

display(daily_returns.head())
```

```python
import statsmodels.api as sm

# Define the dependent and independent variables
Y = daily_returns['BTC-USD']
X = daily_returns['CL=F']

# Add a constant (itercept) to the independent variable
X = sm.add_constant(X)

# Perform OLS regression
model = sm.OLS(Y, X).fit()

# Display the regression rsults
print(model.summary())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Initialize and train the Random Forrest Regressor model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, Y_train)

# Make predictions on the test set
Y_pred = model.predict(X_test)

# Calculate performance metrics
mae = mean_absolute_error(Y_test, Y_pred)
rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

# Task
Run a Random Forest regression model to predict 'BTC-USD' using 'CL=F' as the predictor. Split the data into 80% training and 20% testing sets. Calculate and report the RMSE and MAE of the model.

## Prepare data

### Subtask:
Ensure the dependent and independent variables (`Y` and `X`) are in the correct format for scikit-learn models.


**Reasoning**:
Convert the pandas Series and DataFrame to NumPy arrays to prepare them for scikit-learn models.



```python
Y_array = Y.values
X_array = X.values
```

## Split data

### Subtask:
Split the `X` and `Y` data into training and testing sets using an 80/20 ratio.


**Reasoning**:
Split the data into training and testing sets using an 80/20 ratio.



```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_array, Y_array, test_size=0.2, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of Y_train:", Y_train.shape)
print("Shape of Y_test:", Y_test.shape)
```

## Train model

### Subtask:
Train a Random Forest Regressor model on the training data.


**Reasoning**:
Import the necessary class and instantiate and train the Random Forest Regressor model using the training data.



```python
from sklearn.ensemble import RandomForestRegressor

# Instantiate a RandomForestRegressor object
rf_model = RandomForestRegressor(random_state=42)

# Fit the model to the training data
rf_model.fit(X_train, Y_train)
```

## Make predictions

### Subtask:
Use the trained model to make predictions on the testing data.


**Reasoning**:
Use the trained Random Forest model to make predictions on the test set.



```python
Y_pred = rf_model.predict(X_test)
```

## Evaluate model

### Subtask:
Calculate and print the Root Mean Squared Error (RMSE) and Mean Absolute Error (MAE) of the model's predictions.


**Reasoning**:
Calculate and print the RMSE and MAE of the model's predictions.



```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))

# Calculate MAE
mae = mean_absolute_error(Y_test, Y_pred)

# Print the results
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
```
